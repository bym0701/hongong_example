```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
print(diabetes.data.shape)
print(diabetes.data)
```

```python
print(diabetes.feature_names)
```

```python
print(diabetes.target.shape)
print(diabetes.target)
```

```python
#age와의 상관관계
diabetes_age = diabetes.data[:, 0]
print(diabetes_age)
```

```python
diabetes_age = diabetes.data[:, np.newaxis, 0]
print(diabetes_age)
```

```python
diabetes_target = diabetes.target[:, np.newaxis]
print(diabetes_target.shape)
```

```python
plt.scatter(diabetes_age, diabetes_target)
plt.xlabel('age')
plt.ylabel('diabetes')
plt.show()
```

```python
#bmi와의 상관관계
diabetes_bmi = diabetes.data[:, np.newaxis, 2]
print(diabetes_bmi.shape)
print(diabetes_bmi)

plt.scatter(diabetes_bmi, diabetes_target)
plt.xlabel('bmi')
plt.ylabel('diabetes')
plt.show()
```

```python
#bp와의 상관관계
diabetes_bp = diabetes.data[:, np.newaxis, 3]
print(diabetes_bp)

plt.scatter(diabetes_bp, diabetes_target)
plt.xlabel('bp')
plt.ylabel('diabetes')
plt.show()
```

```python
#bmi로 회귀
from sklearn.model_selection import train_test_split

train_bmi, test_bmi, train_target, test_target = train_test_split(diabetes_bmi, diabetes_target, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(train_bmi, train_target)
```

```python
print(lr.coef_, lr.intercept_)
```

```python
print(lr.score(train_bmi, train_target))
print(lr.score(test_bmi, test_target))
```

```python
#bmi 다항회귀
train_bmi_poly = np.column_stack((train_bmi ** 2, train_bmi))
test_bmi_poly = np.column_stack((test_bmi ** 2, test_bmi))
```

```python
print(train_bmi_poly.shape, test_bmi_poly.shape)
```

```python
lr.fit(train_bmi_poly, train_target)

print(lr.score(train_bmi_poly, train_target))
print(lr.score(test_bmi_poly, test_target))
```

```python
#bp로 회귀

train_bp, test_bp, train_target, test_target = train_test_split(diabetes_bp, diabetes_target, test_size=0.2)

lr.fit(train_bp, train_target)
print(lr.score(train_bp, train_target))
print(lr.score(test_bp, test_target))
```

```python
#bp 다항회귀
train_bp_poly = np.column_stack((train_bp ** 2, train_bp))
test_bp_poly = np.column_stack((test_bp ** 2, test_bp))

print(train_bmi_poly.shape, test_bmi_poly.shape)

lr.fit(train_bp_poly, train_target)

print(lr.score(train_bp_poly, train_target))
print(lr.score(test_bp_poly, test_target))
```
